```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def Euler(y0,t0,dt=.1,n=10):
    ypath = [y0]
    t = [t0+j*dt for j in range(0,n+1)]
    for j in range(0,n):
        dy = f(t[j],ypath[-1])*dt
        ypath.append(ypath[-1]+dy)
        #print(dy/dt)
    return [t,ypath]
def RK4(y0,t0,dt = .1, n=10):
    ypath=[y0]
    t = [t0+j*dt for j in range(0,n+1)]
    for j in range(0,n):
        tn = t[j]
        yn = ypath[-1]
        k1 = f(tn,yn)
        k2 = f(tn+dt/2,yn+dt*k1/2)
        k3 = f(tn+dt/2,yn+dt*k2/2)
        k4 = f(tn+dt,yn+dt*k3)
        ypath.append(yn+(k1+2*k2+2*k3+k4)*dt/6)
    return [t,ypath]
```

```python
def f(t,y):
    return t**2-1+2*y
n=float(5)
tend = 1
[T,y] = Euler(1,0,tend/n,int(n))
[T1,y1] = RK4(1,0,tend/n,int(n))
T_new = [0+tend*j/1000 for j in range(0,1001)]
y_exact = [(-2*t**2-2*t+3*np.exp(2*t)+1)/4 for t in T_new]
#plt.plot(T,y,'r',T1,y1,'b',T_new,y_exact,'k')
plt.plot(T,y,'r',T_new,y_exact,'k')
```
